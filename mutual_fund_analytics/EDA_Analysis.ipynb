```python
import sqlite3
import pandas as pd
import plotly.express as px

# Connect to the local SQLite database
conn = sqlite3.connect("bluestock_mf.db")
df_nav = pd.read_sql_query("SELECT * FROM fact_nav", conn)
df_nav['date'] = pd.to_datetime(df_nav['date'])

# Aggregate and format daily trends
df_daily_nav = df_nav.groupby('date')['nav'].mean().reset_index()
fig = px.line(df_daily_nav, x='date', y='nav', 
              title='Daily NAV Trend Analysis (2022 - 2026)',
              labels={'date': 'Timeline', 'nav': 'Average NAV'})

# Highlight 2023 Bull Run and 2024 Market Correction windows
fig.add_vrect(x0="2023-01-01", x1="2023-12-31", fillcolor="green", opacity=0.1, annotation_text="2023 Bull Run")
fig.add_vrect(x0="2024-01-01", x1="2024-06-30", fillcolor="red", opacity=0.1, annotation_text="2024 Corrections")

fig.update_layout(template="plotly_dark")
fig.show()
```
